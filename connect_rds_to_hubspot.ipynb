```python
def send_hubspot_events(df, hubspot_write_key, event_type='contact'):
    """
    Input is a dataframe with anonymous_id in it. If no anonymous_id, error is raised.
    It converts the dataframe into a list of dictionaries and sends events to hubspot.
    arguments:
        df: Input DataFrame.
        hubspot_write_key: Key to communicate with hubspot source.
        event_type: By default it is set to 'contact' indicating that it will transfer data into Contacts.
        In order to transfer data into Deal, set it to 'deal'.
    """

    import requests
    import pandas as pd

    assert 'anonymous_id' in list(df.columns)
    link = 'https://api.hubapi.com/contacts/v1/'+event_type+'/'

    i = 0
    max_queue_size = 10000
    if event_type not in ('contact', 'deal'):
        raise Exception('event_type must be one of: contact, deal')

    form = df.set_index('anonymous_id')
    contact_form = {}

    for index,row in zip(form.index,form.values):
        i += 1
        properties = []
        for a,colname in enumerate(form.columns):
            properties = properties + [dict([("property",colname),("value",row[a])])]
            contact_form = dict([("anonymous_id",index),("properties",properties)])
            requests.post(link, json=contact_form, params={'hapikey': hubspot_write_key})
        if i % max_queue_size == 0:
            raise Exception('max_queue_size exceeded')
```

```python

def send_hubspot_events(df, hubspot_write_key, event_type='contact'):
    """
    Input is a dataframe with anonymous_id in it. If no anonymous_id, error is raised.
    It converts the dataframe into a list of dictionaries and sends events to hubspot.
    arguments:
        df: Input DataFrame.
        hubspot_write_key: Key to communicate with hubspot source.
        event_type: By default it is set to 'contact' indicating that it will transfer data into Contacts.
        In order to transfer data into Deal, set it to 'deal'.
    """

    import requests
    import pandas as pd

    if event_type not in ('contact', 'deal update'):
        raise Exception('event_type must be one of: contact, deal update')

    if event_type == 'contact':
        
        assert 'anonymous_id' in list(df.columns)
        link = 'https://api.hubapi.com/contacts/v1/contact/'

        i = 0
        max_queue_size = 10000

        form = df.set_index('anonymous_id')
        contact_form = {}

        for index, row in zip(form.index, form.values):
            i += 1
            properties = []
            for a, colname in enumerate(form.columns):
                properties = properties + [dict([("property", colname), ("value", row[a])])]
            contact_form = dict([("anonymous_id", index), ("properties", properties)])
            requests.post(link, json=contact_form, params={'hapikey': hubspot_write_key})

            if i % max_queue_size == 0:
                raise Exception('max_queue_size exceeded')

    elif event_type == 'deal update':
        
        assert 'email' in list(df.columns)
        
        i = 0
        max_queue_size = 10000

        # get all deals
        get_deal_link = 'https://api.hubapi.com/deals/v1/deal/paged?hapikey='+hubspot_write_key\
                       +'&includeAssociations=true&properties=dealname'
        get_deal = requests.get(get_deal_link)

        form = df.set_index('email')

        for index,row in zip(form.index,form.values):
            i += 1

            # get vid
            get_vid_link = 'https://api.hubapi.com/contacts/v1/contact/emails/batch/?email='+index\
                          +'&hapikey='+hubspot_write_key
            get_vid = requests.get(get_vid_link)
            vid = list(get_vid.json().keys())[0]     

            # get deal_id
            deal_id = ''
            for k in range(len(get_deal.json()['deals'])):
                try:
                    if str(get_deal.json()['deals'][k]['associations']['associatedVids'][0]) == vid:
                        deal_id = str(get_deal.json()['deals'][k]['dealId'])
                except Exception:
                    pass

            # update_deal_data
            properties = []
            for a,colname in enumerate(form.columns):
                properties = properties + [dict([("name", colname), ("value", row[a])])]
            update_data = dict([("properties", properties)])

            # put data
            put_data_link = 'https://api.hubapi.com/deals/v1/deal/'+deal_id
            requests.put(put_data_link,json=update_data, params={'hapikey': hubspot_write_key})

            if i % max_queue_size == 0:
                raise Exception('max_queue_size exceeded')

```
